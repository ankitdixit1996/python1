1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# Visitng the site url
try :
    url1='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
    driver.get(url1)
except WebDriverException :
    print('WebDriverException')
```

```python
# Creating empty lists
Rank=[]
Name=[]
Artist=[]
U_date=[]
Views=[]
```

```python
# Scraping details

try :
    rn=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in rn :
        r=i.text
        Rank.append(r)
except NoSuchElementException :
    Rank.append('-')

time.sleep(1)
try :
    nm=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in nm :
        n=i.text.split('[')[0]
        Name.append(n)
except NoSuchElementException :
    Name.append('-')
    
time.sleep(1)
try :
    ar=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in ar :
        a=i.text
        Artist.append(a)
except NoSuchElementException :
    Artist.append('-')

time.sleep(1)
try :
    ud=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
    for i in ud :
        u=i.text
        U_date.append(u)
except NoSuchElementException :
    U_date.append('-')
    
time.sleep(1)
try :
    vw=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in vw :
        v=i.text
        Views.append(v)
except NoSuchElementException :
    Views.append('-')
```

```python
# Printing lengths
print(len(Rank),len(Name),len(Artist),len(U_date),len(Views))
```

```python
# Make dataframe
df1 = pd.DataFrame({'Rank':Rank,'Name':Name,'Artist':Artist,'U_date':U_date,'Views':Views})
df1
```
