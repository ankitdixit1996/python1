### Steps and assumptions with Chi-square test

```python
import scipy.stats as stats

observed1=pd.crosstab(df['1Gender of respondent'], df['3 Which city do you shop online from?'])

chi2,p,dof,expected=stats.chi2_contingency(observed=observed1)
print('Chi square test : ',chi2)
print('p-value : ',p)
print('Dof : ',dof)
print('Expected : ',expected)
```

p-value=3.9 i.e greater than 0.05(Significant value).
Therefore, Ho is accepted and shows two variables are independent of each other.

```python
observed2=pd.crosstab(df['1Gender of respondent'],df['2 How old are you? '])

chi2,p,dof,expected=stats.chi2_contingency(observed=observed2)
print('Chi square test : ',chi2)
print('p-value : ',p)
print('Dof : ',dof)
print('Expected : ',expected)
```
